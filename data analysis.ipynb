```python
numerical_df = dataset.select_dtypes(include=["int64", "float64"])
```

```python
dataset.head()
```

```python
def calc_stats_metrics(numerical_df):
    metrics = {
        'Range': [],
        'Mean': [],
        'Mode': [],
        'Standard Deviation': [],
        'IQR': []
    }
    
    for attribute in numerical_df.columns:
        # Calculating range
        attr_min = numerical_df[attribute].min()
        attr_max = numerical_df[attribute].max()
        metrics['Range'].append(f"[{attr_min}, {attr_max}]")
        
        # Calculating mean
        metrics['Mean'].append(round(numerical_df[attribute].mean(), 6))
        
        # Calculating mode
        metrics['Mode'].append(numerical_df[attribute].mode()[0])
        
        # Calculating standard deviation
        metrics['Standard Deviation'].append(round(numerical_df[attribute].std(), 6))
        
        # Calculating Inter Quartile Range (IQR)
        Q1 = numerical_df[attribute].quantile(0.25)
        Q3 = numerical_df[attribute].quantile(0.75)
        IQR = Q3 - Q1
        metrics['IQR'].append(round(IQR, 6))
    
    return pd.DataFrame(metrics, index=numerical_df.columns)
```

```python
numerical_stats = calc_stats_metrics(numerical_df)
```

```python
numerical_stats.T
```

```python
corr_matrix = numerical_df.corr()

# Set the plot size for better visibility
plt.figure(figsize=(12, 10))

# Create a heatmap with better color schemes and annotations for clarity
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f', annot_kws={"size": 10})

# Set the title for the heatmap
plt.title("Correlation Heatmap of Numerical Features", fontsize=16)

# Save the heatmap as an image file
plt.savefig('correlation_heatmap_numerical_features.png', dpi=300, bbox_inches='tight')

# Display the heatmap
plt.show()
```

```python
important_features = [
    'credit_history_age', 'num_credit_card', 'interest_rate', 
    'num_credit_inquiries', 'num_bank_accounts', 'outstanding_debt', 
    'annual_income', 'monthly_balance', 'credit_score'
]

# Filter the dataframe to only include the important features
important_features_df = dataset[important_features]

# Calculate the correlation matrix for the important features
important_corr_matrix = important_features_df.corr()

# Set the plot size for better visibility
plt.figure(figsize=(10, 8))

# Create a heatmap with better color schemes and annotations for clarity
sns.heatmap(important_corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f', annot_kws={"size": 10})

# Set the title for the heatmap
plt.title("Correlation Heatmap of Important Features", fontsize=16)

plt.xticks(rotation=45)

# Display the heatmap
plt.show()
```

```python
# List of selected features for histograms
histogram_features = ['interest_rate', 'credit_history_age', 'outstanding_debt']

# Set the plot size for better visibility
plt.figure(figsize=(15, 5))

# Loop through the list of features and plot histograms
for i, feature in enumerate(histogram_features, 1):
    plt.subplot(1, 3, i)  # Create a 1x3 grid of subplots
    sns.histplot(dataset[feature], kde=True)
    plt.title(f"Distribution of {feature.replace('_', ' ').title()}")
    plt.xlabel(feature.replace('_', ' ').title())
    plt.ylabel('Frequency')


plt.suptitle('Feature Distributions: Interest Rate, Credit History Age, and Outstanding Debt', fontsize=16)

# Adjust layout for better spacing between subplots
plt.tight_layout()

# Save the plot
plt.savefig('selected_feature_histograms.png', dpi=300)

# Display the plot
plt.show()
```

```python
# Selected features for pair plot
pair_plot_features = ['credit_score', 'credit_history_age', 
                      'interest_rate', 'outstanding_debt']

# Create the pair plot
sns.pairplot(dataset[pair_plot_features], hue='credit_score')

# Set the title for the entire plot (adjust the layout to fit the title if necessary)
plt.suptitle("Pair Plot of Highly Correlated Features", y=1.02, fontsize=16)

plt.savefig('selected_feature_pairplot.png', dpi=300)

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(18, 12)) 

# Selected features for a less congested pair plot
reduced_pair_plot_features = ['credit_score', 'credit_history_age', 'interest_rate', 'outstanding_debt']

# Create the pair plot with reduced features
pairplot = sns.pairplot(dataset[reduced_pair_plot_features], hue='credit_score', palette='coolwarm')

# Set the title for the entire plot
plt.suptitle("Pair Plot of Key Features: Credit Score, Credit History, Interest Rate, and Outstanding Debt", 
             y=1.03, fontsize=16)

# Adjust the layout to prevent overlap and make space for the legend
plt.subplots_adjust(top=0.92, right=0.8)  # Adjust the right side to make space for the legend

# Move the legend outside the plot
pairplot._legend.set_bbox_to_anchor((1.05, 0.5))  # Legend to the right, centered vertically
pairplot._legend.set_title("Credit Score")

# Save the plot with better bounding box options
pairplot.savefig('reduced_feature_pairplot.png', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(10, 5))

# 1. Box plot for Interest Rate vs Credit Score
plt.subplot(1, 2, 1)
sns.boxplot(x='credit_score', y='interest_rate', data=dataset, palette='Set2', hue="credit_score", legend=False)
plt.title('Interest Rate by Credit Score')
plt.xlabel('Credit Score')
plt.ylabel('Interest Rate')

# 2. Box plot for Outstanding Debt vs Credit Score
plt.subplot(1, 2, 2)
sns.boxplot(x='credit_score', y='outstanding_debt', data=dataset, palette='Set2', hue="credit_score", legend=False)
plt.title('Outstanding Debt by Credit Score')
plt.xlabel('Credit Score')
plt.ylabel('Outstanding Debt')

# Adjust layout for better spacing between the two plots
plt.tight_layout()

# Save the plot as a square image
plt.savefig('boxplot_interest_outstanding_square.png', dpi=300)

# Display the plot
plt.show()
```

```python
# Set the size of the plot
plt.figure(figsize=(10, 6))

# Create the bar plot
sns.countplot(data=dataset, x='payment_behaviour', hue='credit_score', palette='Set2')

# Set the title and labels
plt.title('Trends in Payment Behaviors and Credit Scores', fontsize=16)
plt.xlabel('Payment Behaviour', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Rotate x labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()

plt.savefig('bar_plot_trends.png', dpi=300)
plt.show()
```
